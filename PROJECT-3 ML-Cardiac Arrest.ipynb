```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(5)
```

```python
model.fit(cc)
```

```python
model.predict(cc)
```

```python
cc["cluster"]=model.predict(cc)
```

```python
cc
```

```python
df1=cc[cc["cluster"]==0]
df2=cc[cc["cluster"]==1]
df3=cc[cc["cluster"]==2]
df4=cc[cc["cluster"]==3]
df5=cc[cc["cluster"]==4]
```

```python
plt.scatter(df1,df1,color="r")
plt.scatter(df2,df2,color="b")
plt.scatter(df3,df3,color="purple")
plt.scatter(df4,df4,color="y")
plt.scatter(df5,df5,color="g")
```
